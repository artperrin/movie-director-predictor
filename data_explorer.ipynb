*This script allows the user to access informations about his training (and testing) dataset.*

```python
import matplotlib.pyplot as plt
import matplotlib
import os

path = './data_trn/'
```

```python
# searching for the directories names and how many files are in each directory
directories_names = []
for _, directories, _ in os.walk(path, topdown=False):
    for name in directories:
        directories_names.append(name)

files_per_directories = []
for name in directories_names:
    Npath = os.path.join(path, name)
    files_per_directories.append(len(os.listdir(Npath)))
```

```python
# plot a bar diagram
plt.style.use("ggplot")
plt.figure(figsize=(25,7))
plt.bar(directories_names, files_per_directories)
plt.xlabel('director')
plt.ylabel('Number of random screenshots')
plt.title('Training dataset')
plt.show()
```
